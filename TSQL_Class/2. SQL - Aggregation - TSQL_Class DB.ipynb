```python
import pyodbc
import pandas as pd

import warnings
warnings.filterwarnings('ignore') 

# defining the server and the database
server = '(localdb)\mssqllocaldb' 
database = 'TSQL_CLASS'  

# Define the connection string
conn = pyodbc.connect(
    'DRIVER={ODBC Driver 17 for SQL Server}; \
    SERVER='+ server +'; \
    DATABASE ='+ database +';\
    Trusted_Connection=yes;'
) 
```

### Question 1 - TSQL_Class - Products table
### 1. Compute the average price of all the products.

```python
# Using a TOP 5 filter to limit the amount of data pulled from exploratory query
eda_query = pd.read_sql_query("""SELECT TOP 10 * FROM [TSQL_Class].[dbo].[Products]""",conn)
eda_query

```

```python
# Calculate the average price of all the products in Products table using the AVG() function 
query = pd.read_sql_query("""SELECT AVG(Price) AS avg_price 
                             FROM [TSQL_CLASS].[dbo].[Products]""",conn)
query
```

### Question 4 - TSQL_Class DB - Products table
### 4. Select the average price of each manufacturer's

### Question 8 - TSQL_Class DB - Boxes table
### 8. Select the warehouse and the average value of the

### Question 9 - TSQL_Class DB - Boxes table
### 9. Same as previous exercise, but select only those warehouses 
### where the average value of the boxes is greater than 150.

```python
# Using a TOP 10 filter to limit the amount of data pulled from exploratory query
eda_query = pd.read_sql_query("""SELECT TOP 10 * FROM [TSQL_Class].[dbo].[Boxes]""",conn)
eda_query
```

```python
# Using Having statement to filter for warehouse with avg value greater than to equal to 150
sql_query = pd.read_sql_query("""SELECT Warehouse, AVG(VALUE) AS avg_value FROM [TSQL_Class].[dbo].[Boxes]
                                 GROUP BY Warehouse
                                 HAVING AVG(VALUE) >= 150""",conn)
sql_query
```
