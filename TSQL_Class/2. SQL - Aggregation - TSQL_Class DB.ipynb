```python
import pyodbc

# defining the server and the database
server = '(localdb)\mssqllocaldb' 
database = 'TSQL_CLASS'  

# Define the connection string
conn = pyodbc.connect(
    'DRIVER={ODBC Driver 17 for SQL Server}; \
    SERVER='+ server +'; \
    DATABASE ='+ database +';\
    Trusted_Connection=yes;'
)
```

```python
# Question 1 - TSQL_Class - Products table
# Compute the average price of all the products.

# Create the Cursor.
cursor = conn.cursor()

# Using a TOP 5 filter to limit the amount of data pulled from exploratory query
cursor.execute("""SELECT TOP 5 * FROM [TSQL_Class].[dbo].[Products]""")

# Using a for loop to print a of the column headers for table
column_headers = [i[0] for i in cursor.description]
print(column_headers)

sql_query = cursor.fetchall()
for row in sql_query:
    print(row)
    
print("\n")

# Calculate the average price of all the products in Products table using the AVG() function 
cursor.execute("""SELECT AVG(Price) AS avg_price FROM [TSQL_CLASS].[dbo].[Products]""")

column_headers = [i[0] for i in cursor.description]
print(column_headers)

sql_query = cursor.fetchall()
for row in sql_query:
    print(row)
    
# close the cursor
# cursor.close()
# conn.close()
```

```python
# Question 9 - TSQL_Class DB - Boxes table
# Same as previous exercise, but select only those warehouses 
# where the average value of the boxes is greater than 150.

# Create the Cursor
cursor = conn.cursor()

# Using a TOP 10 filter to limit the amount of data pulled from exploratory query
cursor.execute("""SELECT TOP 10 * FROM [TSQL_Class].[dbo].[Boxes]""")

# Using a for loop to print a of the column headers for table
column_headers = [i[0] for i in cursor.description]
print(column_headers)

sql_query = cursor.fetchall()
for row in sql_query:
    print(row)
    
print("\n")

# Using Having statement to filter for warehouse with avg value greater than to equal to 150
cursor.execute("""SELECT Warehouse, AVG(VALUE) AS avg_value FROM [TSQL_Class].[dbo].[Boxes]
GROUP BY Warehouse
HAVING AVG(VALUE) >= 150""")

column_headers = [i[0] for i in cursor.description]
print(column_headers)

sql_query = cursor.fetchall()
for row in sql_query:
    print(row)
    
# close the cursor
# cursor.close()
# conn.close()
```
