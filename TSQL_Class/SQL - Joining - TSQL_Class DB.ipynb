```python
# Question 5 - TSQL_Class - Products table & Manufacturing table
# Select the names of manufacturer whose products 
# have an average price larger than or equal to $150


# Create the Cursor.
cursor = conn.cursor()

# Using a TOP 5 filter to limit the amount of data pulled from exploratory query
cursor.execute("""SELECT TOP 5 * FROM [TSQL_Class].[dbo].[Products]""")

# Using a for loop to print a of the column headers for table
column_headers = [i[0] for i in cursor.description]
print(column_headers)

sql_query = cursor.fetchall()
for row in sql_query:
    print(row)
    
print("\n")

# Using a TOP 5 filter to limit the amount of data pulled from exploratory query
cursor.execute("""SELECT TOP 5 * FROM [TSQL_Class].[dbo].[Manufacturers]""")

# Using a for loop to print a of the column headers for table
column_headers = [i[0] for i in cursor.description]
print(column_headers)

sql_query = cursor.fetchall()
for row in sql_query:
    print(row)
    
print("\n")

# Using Select with to get Average Price column from the products table including 
# Name column from manufacturer table. Adding a HAVING clause to filter Mfg names with avgs >= 150 
cursor.execute("""SELECT AVG(Prod.Price) AS Average, Mfg.Name FROM [TSQL_CLASS].[dbo].[Products] AS Prod
                FULL JOIN [TSQL_CLASS].[dbo].[Manufacturers] AS Mfg
                ON Prod.Manufacturer = Mfg.Code
                GROUP BY Mfg.Name
                HAVING AVG(Prod.Price) >= 150""")

column_headers = [i[0] for i in cursor.description]
print(column_headers)

sql_query = cursor.fetchall()
for row in sql_query:
   print(row)
    
# close the cursor
# cursor.close()
# conn.close()
```

```python
# Question 10 - TSQL_Class - Pieces, Provides and Providers table
# Obtain the names of all providers who supply piece 1.


# Create the Cursor.
cursor = conn.cursor()

# Using a TOP 5 filter to limit the amount of data pulled from exploratory query
cursor.execute("""SELECT TOP 5 * FROM [TSQL_Class].[dbo].[Pieces]""")

# Using a for loop to print a of the column headers for table
column_headers = [i[0] for i in cursor.description]
print(column_headers)

sql_query = cursor.fetchall()
for row in sql_query:
    print(row)
    
print("\n")

# Using a TOP 5 filter to limit the amount of data pulled from exploratory query
cursor.execute("""SELECT TOP 5 * FROM [TSQL_Class].[dbo].[Provides]""")

# Using a for loop to print a of the column headers for table
column_headers = [i[0] for i in cursor.description]
print(column_headers)

sql_query = cursor.fetchall()
for row in sql_query:
    print(row)
    
print("\n")

# Using a TOP 5 filter to limit the amount of data pulled from exploratory query
cursor.execute("""SELECT TOP 5 * FROM [TSQL_Class].[dbo].[Providers]""")

# Using a for loop to print a of the column headers for table
column_headers = [i[0] for i in cursor.description]
print(column_headers)

sql_query = cursor.fetchall()
for row in sql_query:
    print(row)
    
print("\n")


# Using Full Join to combine three tables Pieces, Provides, and Providers
# Using SELECT to specify which columns to pull Provider.Name which provides piece 1 
cursor.execute("""SELECT Providers.Name FROM [TSQL_Class].[dbo].[Pieces] AS Pieces
                  FULL JOIN [TSQL_Class].[dbo].[Provides] AS Provides
                  ON Pieces.Code = Provides.Piece
                  FULL JOIN [TSQL_Class].[dbo].[Providers] AS Providers
                  ON Providers.Code = Provides.Provider
                  WHERE Pieces.Code = 1""")

column_headers = [i[0] for i in cursor.description]
print(column_headers)

sql_query = cursor.fetchall()
for row in sql_query:
   print(row)
    
# close the cursor
# cursor.close()
# conn.close()
```
