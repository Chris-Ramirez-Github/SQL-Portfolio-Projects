```python
# Using Select with to get Average Price column from the products table including 
# Name column from manufacturer table. Adding a HAVING clause to filter Mfg names with avgs >= 150 
join_query = pd.read_sql_query("""SELECT AVG(Prod.Price) AS [Average Price], Mfg.Name 
                                  FROM [TSQL_CLASS].[dbo].[Products] AS Prod
                                  FULL JOIN [TSQL_CLASS].[dbo].[Manufacturers] AS Mfg
                                  ON Prod.Manufacturer = Mfg.Code
                                  GROUP BY Mfg.Name
                                  HAVING AVG(Prod.Price) >= 150""",conn)
join_query
```

```python
# Using Full Join to combine three tables Pieces, Provides, and Providers
# Using SELECT to specify which columns to pull Provider.Name which provides piece 1 
join_query = pd.read_sql_query("""SELECT Providers.Name FROM [TSQL_Class].[dbo].[Pieces] AS Pieces
                                  FULL JOIN [TSQL_Class].[dbo].[Provides] AS Provides
                                  ON Pieces.Code = Provides.Piece
                                  FULL JOIN [TSQL_Class].[dbo].[Providers] AS Providers
                                  ON Providers.Code = Provides.Provider
                                  WHERE Pieces.Code = 1""",conn)
join_query
```

```python
# Using Full Join to combine three tables Pieces, Provides, and Providers
# Using SELECT to specify which columns to pull in this case Pieces.Name which are provided 
# by Provider "HAL"
# Primary Key: Providers.Code = Provides.Provider, Pieces.Code = Provider.Piece
join_query = pd.read_sql_query("""SELECT Pieces.Name FROM [TSQL_Class].[dbo].[Providers] AS Providers
                                  FULL JOIN [TSQL_Class].[dbo].[Provides] AS Provides
                                  ON Providers.Code = Provides.Provider
                                  FULL JOIN [TSQL_Class].[dbo].[Pieces] AS Pieces
                                  ON Pieces.Code = Provides.Piece
                                  WHERE Providers.Code = 'HAL'""",conn)
join_query
```
