### Question 15 - AdventureWorks DB - HumanResources.Employee table
### 15. List the dates where new employees were hired. 
### • Make sure you list them from the recent to oldest
### • Make sure no repeating dates in the result

```python
# Using SELECT to pull hiring dates for new employees
# Using ORDER BY to list the hiring dates from recent to oldest and DISTINCT 
# Using DESC with ORDER BY to sort list from recent to oldest
sql_query = pd.read_sql_query("""SELECT DISTINCT Emp.HireDate
                                 FROM [AdventureWorks2012].[HumanResources].[Employee] AS Emp
                                 ORDER BY Emp.HireDate DESC""",conn)
sql_query
```

### Question 16 - AdventureWorks DB - Production.Product table
### 16. Get the last 5 persons which records where modified (modifiedDates) most recently

```python
# Using DESC in Order by to get the most recent records for modifiedDate in Person table
# Limiting results by using TOP 5
sql_query = pd.read_sql_query("""SELECT TOP 5 Person.FirstName, Person.LastName, Person.modifiedDate
                                 FROM [AdventureWorks2012].[Person].[Person] AS Person
                                 ORDER BY Person.modifiedDate DESC""",conn)
sql_query
```

### Question 17 - AdventureWorks DB - Person.Person table
### 17.Repeat the previous question however, in this case, if the 5th
### person’s record was modified at the same time as other people’s 
### records, get all the pertinent people

```python
# Using DESC in Order by to get the most recent records for modifiedDate in Person table
# Using WITH TIES clause to get records ties to 5th Person.modifiedDate
# Limiting results by using TOP 5
sql_query = pd.read_sql_query("""SELECT TOP 5 WITH TIES Person.FirstName, Person.LastName, Person.modifiedDate
                                 FROM [AdventureWorks2012].[Person].[Person] AS Person
                                 ORDER BY Person.modifiedDate DESC""",conn)
sql_query
```
